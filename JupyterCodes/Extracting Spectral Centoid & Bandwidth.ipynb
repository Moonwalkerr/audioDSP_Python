```python
import matplotlib.pyplot as plt
import numpy as np
import librosa
import IPython.display as ipd
```

```python
debussy_file = "audios/debussy.wav"
redhot_file = "audios/redhot.wav"
duke_file = "audios/duke.wav"
```

```python
## Loading audio files with librosa

debussy , sr = librosa.load(debussy_file)
redhot, _ = librosa.load(redhot_file)
duke, _ = librosa.load(duke_file)
```

## Extracting Spectral Centroid with librosa

```python
FRAME_SIZE = 1024
HOP_LENGTH = 512
```

```python
sc_debussy = librosa.feature.spectral_centroid(y=debussy, sr=sr,n_fft=FRAME_SIZE, hop_length=HOP_LENGTH )
sc_debussy.shape
```

```python
sc_redhot = librosa.feature.spectral_centroid(y=redhot, sr=sr,n_fft=FRAME_SIZE, hop_length=HOP_LENGTH )
sc_duke = librosa.feature.spectral_centroid(y=duke, sr=sr,n_fft=FRAME_SIZE, hop_length=HOP_LENGTH )
```

```python
sc_redhot.shape
```

```python
## 2nd dim is no of frames
sc_duke.shape
```

```python
## But we are mostly intrested in no of frames, i.e 2nd dim, so

sc_debussy = librosa.feature.spectral_centroid(y=debussy, sr=sr,n_fft=FRAME_SIZE, hop_length=HOP_LENGTH )[0]
sc_redhot = librosa.feature.spectral_centroid(y=redhot, sr=sr,n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
sc_duke = librosa.feature.spectral_centroid(y=duke, sr=sr,n_fft=FRAME_SIZE, hop_length=HOP_LENGTH )[0]
```

```python
## Now seems good

sc_debussy.shape
```

## Visualizting SC across time

```python
plt.figure(figsize=(25,10))

frames = range(len(sc_debussy))
t = librosa.frames_to_time(frames, hop_length=HOP_LENGTH)
plt.plot(t, sc_debussy, color="b")
plt.plot(t, sc_redhot, color="r")
plt.plot(t, sc_duke, color="g")

```

## Extracting Spectral Bandwidth

```python
bw_debussy = librosa.feature.spectral_bandwidth(y=debussy,sr=sr,n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
bw_redhot = librosa.feature.spectral_bandwidth(y=redhot,sr=sr,n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
bw_duke = librosa.feature.spectral_bandwidth(y=duke,sr=sr,n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
```

```python
bw_debussy.shape
```

## Visualizing Bandwidth 

```python
plt.figure(figsize=(25,10))

frames = range(len(bw_debussy))
t = librosa.frames_to_time(frames, hop_length=HOP_LENGTH)

plt.plot(t,bw_debussy, color="b")
plt.plot(t,bw_redhot, color='r')
plt.plot(t, bw_duke, color='g')
```

### The Bandwidth curve is similar to Spectral curve as it is derived from the spectral centroid
